```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Isi kembali DataFrame dengan data
df = pd.read_csv("/content/drive/My Drive/csv/energy.csv", encoding='unicode_escape')

# Set index dan ubah ke tipe data timestamp
df = df.set_index('timestamp')
df.index = pd.to_datetime(df.index)

# Plot
df.plot(style='.',
        figsize=(15, 5),
        color=sns.color_palette()[0],
        title='Penggunaan Listrik dan Suhu pada Tahun 2012 Sampai 2014')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/csv/energy.csv", encoding='unicode_escape',
parse_dates=True, index_col="timestamp")

df
```

```python
# @title load

from matplotlib import pyplot as plt
df['load'].plot(kind='line', figsize=(8, 4), title='load')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Assuming df is your DataFrame
sns.set(style="whitegrid") # Setting the style to whitegrid for a clean background

plt.figure(figsize=(12, 6)) # Setting the figure size
sns.lineplot(data=df, x='timestamp', y='load', color='blue')

# Adding labels and title
plt.xlabel('timestamp')
plt.ylabel('load')
plt.title('Penggunaan Listrik dan Suhu dari Tahun 2012 Sampai 2014')

plt.show()
```

```python
# Assuming df is your DataFrame with a datetime index
df_resampled = df.resample('M').mean() # Resampling to monthly frequency, using mean as an aggregation function

sns.set(style="whitegrid") # Setting the style to whitegrid for a clean background

# Plotting the 'high' column with seaborn, setting x as the resampled 'Date'
plt.figure(figsize=(12, 6)) # Setting the figure size
sns.lineplot(data=df_resampled, x=df_resampled.index, y='load', color='blue')

# Adding labels and title
plt.xlabel('timestamp')
plt.ylabel('load')
plt.title('Penggunaan Listrik dan Suhu dari Tahun 2012 Sampai 2014')

plt.show()
```
